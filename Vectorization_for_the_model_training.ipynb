#Bag of words



```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({"text":["people watch dswithbrian",
                           "dswithbrian watch dswithbrian",
                           "people write comment",
                           "dswithbrian write comment"],"output":[1,1,0,0]})
df
```

```python
#perform bag of words technique
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df["text"])
```

```python
#vocabulary
print(cv.vocabulary_)
```

```python
bow.toarray()
```

#N Grams

```python
df = pd.DataFrame({"text":["people watch dswithbrian",
                           "dswithbrian watch dswithbrian",
                           "people write comment",
                           "dswithbrian write comment"],"output":[1,1,0,0]})
df
```

```python
#BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
bow = cv.fit_transform(df["text"])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
print(bow[3].toarray())
```

```python
#Ti gram
# Bi grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(3,3))
```

```python
bow = cv.fit_transform(df["text"])
```

```python
print(cv.vocabulary_)
```

TF-IDF(Term frequency-Inverse document frequency)

```python
df = pd.DataFrame({"text":["people watch dswithbrian",
                           "dswithbrian watch dswithbrian",
                           "people write comment",
                           "dswithbrian write comment"],"output":[1,1,0,0]})
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid = TfidfVectorizer()
```

```python
arr = tfid.fit_transform(df["text"]).toarray()
arr
```

```python
print(tfid.idf_)
```
